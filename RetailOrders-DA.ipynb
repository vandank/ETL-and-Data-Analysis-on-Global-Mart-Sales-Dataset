```python
#Extract file from zip file
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall() #extract file to dir
zip_ref.close()
```

```python
#Now read the data from the file and handle the null values
import pandas as pd
df = pd.read_csv('orders.csv')
df.head(20)
```

```python
#In Ship Mode column, we can see some values as Not available, Unknown etc. We will treat this values as Null values
df['Ship Mode'].unique()
```

```python
#In order to do this, we will pass a parameter na_values=['Not Available', 'unknown'] to consider these two values as null values
df = pd.read_csv('orders.csv',na_values=['Not Available','Unknown'])
df.head(20)
df['Ship Mode'].unique()
```

```python
#Now these two values are treated as null values
#Now we'll work on renaming the columns just for our convenience. Make them in lower case and replace the spaces with underscore
#One way of doing this:
df.rename(columns={'Order Id':'order_id','City':'city'})
```

```python
#But this will be a tedious process if there are numerous clmns. We will rather use this approach
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
df
```

```python
#Now deriving new columns like discount, sale_price, profit etc
df['discount'] = df['list_price']*df['discount_percent']*.01 #Total discount on an item
df['sale_price'] = df['list_price'] - df['discount']
df['profit'] = df['sale_price'] - df['cost_price']
df.head(5)
```

```python
#We can look at the order_date column that the data type of the order_date is object. It should rather be date type bcz when we load this in sql server
#It should go as date type. 
df.dtypes
```

```python
#Converting order date from object data type to datetime
df['order_date'] = pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
df.dtypes
```

```python
#Since we now have discount, sale_price and profit, we can do analysis with these 3 columns so we don't require the cost price, and discount_percent columns
#So, we'll drop them
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
df
```
